```python
'''Mempersiapkan Data: Mengimpor data transaksi dan menyiapkan fitur untuk clustering.'''
from sklearn.preprocessing import StandardScaler
import pandas as pd

df = pd.read_csv('data_transaksi.csv')
features = df[['age', 'annual_income', 'spending_score']] # Fitur untuk clustering
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

'''Menerapkan K-Means: Menentukan jumlah cluster dengan Elbow Method dan menjalankan K-Means Clustering.'''
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Menentukan jumlah cluster optimal dengan Elbow Method
wcss = []
for i in range(1, 11): 
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42) 
    kmeans.fit(features_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Jumlah Cluster')
plt.ylabel('WCSS')
plt.show()
```
