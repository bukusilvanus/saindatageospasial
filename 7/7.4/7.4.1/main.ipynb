```python
import pandas as pd
data = pd.read_csv('data.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menggunaakan matplotlib
plt.hist(data['petal_length'], bins=20, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Menggunakan seaborn
sns.histplot(data['petal_length'], bins=20, kde=True, color='green')
plt.title('Histogram with Seaborn')
plt.show()
```

```python
# Menggunakan matplotlib
plt.bar(data['species'].value_counts().index, data['species'].value_counts())
plt.title('Bar Chart') 
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()

# Menggunakan seaborn
sns.countplot(x='species', data=data)
plt.title('Bar Chart with Seaborn')
plt.show()
```

```python
# Menggunakan matplotlib
plt.scatter(data['sepal_length'], data['petal_length'], alpha=0.5, color='red') 
plt.title('Scatter Plot') 
plt.xlabel('Feature 1') 
plt.ylabel('Feature 2') 
plt.show() 

# Menggunakan Seaborn 
sns.scatterplot(x='sepal_length', y='petal_length', data=data, color='purple') 
plt.title('Scatter Plot with Seaborn') 
plt.show() 
```

```python
# Menggunakan Matplotlib 
plt.boxplot(data['sepal_length']) 
plt.title('Box Plot') 
plt.ylabel('Value') 
plt.show() 

# Menggunakan Seaborn 
sns.boxplot(x='species', y='sepal_length', data=data) 
plt.title('Box Plot with Seaborn') 
plt.show() 

# Menggunakan Matplotlib 
plt.scatter(data['sepal_length'], data['petal_length'], alpha=0.5, color='red') 
plt.title('Scatter Plot') 
plt.xlabel('Feature 1') 
plt.ylabel('Feature 2') 
plt.show() 

# Menggunakan Seaborn 
sns.scatterplot(x='sepal_length', y='petal_length', data=data, color='purple') 
plt.title('Scatter Plot with Seaborn') 
plt.show() 
```

```python
plt.plot(data['sepal_length'], data['petal_length'], label='Data Label', color='orange') 
plt.xlabel('Column 1') 
plt.ylabel('Column 2') 
plt.title('Line Plot with Customizations') 
plt.legend(loc='best') 
plt.show() 
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=data, color='blue', s=100, marker='o') 
plt.title('Custom Scatter Plot with Seaborn') 
plt.show() 
```

```python
plt.plot(data['sepal_length'], data['petal_length']) 
plt.grid(True)
plt.xticks(rotation=45) 
plt.yticks(range(0, 101, 10)) 
plt.show() 
```
