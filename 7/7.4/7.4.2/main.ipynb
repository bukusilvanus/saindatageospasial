```python
import geopandas as gpd 
import matplotlib.pyplot as plt 

gdf = gpd.read_file('data.json') 

fig, axes = plt.subplots(1, 1, figsize=(10, 6)) 
gdf['GID_1'].value_counts().plot(kind='bar', ax=axes) 
plt.tight_layout() 

gdf.plot() 
plt.show() 
```
