```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Membaca file shapefile
gdf = gpd.read_file('shapefile.zip')

# Menampilkan data geospasial
print(gdf.head())

# Visualisasi data geospasial
gdf.plot()
plt.show()
```

```python
buffer = gdf.geometry.buffer(0.01)
buffer.plot()

gdf_intersection = gdf.iloc[0].geometry.intersection(buffer)
gdf_union = gdf.iloc[0].geometry.union(buffer)
```

```python
import matplotlib.pyplot as plt

gdf.plot(figsize=(10, 10), edgecolor='black')
plt.title("Visualisasi Data Geospasial")
plt.show()
```

```python
import folium

latitude = gdf.centroid.y.mean()
longitude = gdf.centroid.x.mean()

m = folium.Map(location=[latitude, longitude], zoom_start=12)
folium.Marker([latitude, longitude], popup='Lokasi').add_to(m)
m.save("map.html")
```
