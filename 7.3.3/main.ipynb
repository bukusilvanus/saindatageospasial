## 7.3.3.1

```python
import pandas as pd
data = pd.read_csv('data.csv')
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data[['sepal_length', 'petal_length', 'sepal_width', 'petal_width']])
```

```python
data = pd.DataFrame(data_scaled, columns=['sepal_length', 'petal_length', 'sepal_width', 'petal_width'])
data.head()
```

```python
import pandas as pd
data = pd.read_csv('data.csv')
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
data_scaled = scaler.fit_transform(data[['sepal_length', 'petal_length', 'sepal_width', 'petal_width']])
```
